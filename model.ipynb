# This is reconstruction of LeNet 5 in TensorFlow
This is meant solely as a learning exercise and is neither the fastest nor the best implementation around.

# Imports

```python
from tensorflow.keras import Sequential, layers, Model
from tensorflow.keras.datasets import mnist, cifar10
from tensorflow import pad, expand_dims, cast, float32, int32
from tensorflow.config import list_physical_devices
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
import numpy as np
```

# Base Model
Implemented according to the exact architecture.

```python
class LeNet(Model):
    def __init__(self):
        super().__init__()
        self.conv1 = layers.Conv2D(filters=6, kernel_size=(5, 5), activation='tanh')
        self.pool2 = layers.AveragePooling2D((2, 2), strides=2)
        self.conv3 = layers.Conv2D(filters=16, kernel_size=(5, 5), activation='tanh')
        self.pool4 = layers.AveragePooling2D((2, 2), strides=2)
        self.conv5 = layers.Conv2D(filters=120, kernel_size=(5, 5), activation='tanh')
        self.flat6 = layers.Flatten()
        self.dense7 = layers.Dense(units=84, activation='tanh')
        self.dense8 = layers.Dense(units=10, activation='softmax')

    def call(self, x):
        x = self.conv1(x)
        x = self.pool2(x)
        x = self.conv3(x)
        x = self.pool4(x)
        x = self.conv5(x)
        x = self.flat6(x)
        x = self.dense7(x)
        x = self.dense8(x)

        return x
```

# Testing on MNIST

```python
model = LeNet()
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

## Loading the Dataset

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
padded_x_train = pad(x_train, [[0, 0], [2, 2], [2, 2]], mode='constant', constant_values=0)
padded_x_test = pad(x_test, [[0, 0], [2, 2], [2, 2]], mode='constant', constant_values=0)

padded_x_train = expand_dims(padded_x_train, -1)
padded_x_test = expand_dims(padded_x_test, -1)

padded_x_train.shape
```

## Normalizing

```python
padded_x_train = cast(padded_x_train, float32) / 255.0
padded_x_test = cast(padded_x_test, float32) / 255.0
```

## Training

```python
model.fit(
    padded_x_train,
    y_train,
    epochs=10,
)
```

## Testing

```python
y_pred = model.predict(padded_x_test)
```

```python
y_pred_classes = y_pred.argmax(axis=1)
```

```python
print('MSE:', mean_squared_error(y_test, y_pred_classes))
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_classes)))
print('R2 Score:', r2_score(y_test, y_pred_classes))
print('Accuracy:', accuracy_score(y_test, y_pred_classes))
```
